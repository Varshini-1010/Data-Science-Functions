# *EXCEL FILE OPERATIONS:*

```python
import pandas as pd
students=pd.read_excel("Merge dataset.xlsx",sheet_name='students')
marks=pd.read_excel("Merge dataset.xlsx",sheet_name='marks')
contact=pd.read_excel("Merge dataset.xlsx",sheet_name='contact')
#Inner join
#Features only matching Student_id from both dataframe
inner_df =pd.merge(
    students,
    marks,
    on = 'Student_ID',
    how ='inner'
)
print(inner_df.head())
print(inner_df.shape)
#Left merge
#keeps all records from left dataframe(students)
left_df =pd.merge(
    students,
    marks,
    on = 'Student_ID',
    how ='left'
)
print(left_df.tail())
#Right merge
#keeps all records from right dataframe(students)
right_df =pd.merge(
    students,
    marks,
    on = 'Student_ID',
    how ='right'
)
print(right_df.tail())
#Outer join
#Features matching both dataframe
outer_df =pd.merge(
    students,
    marks,
    on = 'Student_ID',
    how ='outer'
)
print(outer_df.head())
print(outer_df.shape)
#combain students+marks+contacts
student_complete=pd.merge(
    students,
    marks,
    on = 'Student_ID',
    how ='left'
)
student_complete=pd.merge(
    student_complete,
    contact,
    on = 'Student_ID',
    how ='left'
)
print(student_complete.head())

#CONCAT row-wise
#split students into 2 parts
part1=students.iloc[:25]
part2=students.iloc[25:]
combained_rows = pd.concat([part1,part2])
print(combained_rows.shape)

#Concat cloumn-wise
student_basic=students[['Student_ID','Name']]
student_marks=marks[['Python']]
combained_cols=pd.concat(
    [student_basic,student_marks],axis=1)
print(combained_cols.head())

#JOIN
#join() works using index:
student_index=students.set_index('Student_ID')
contact_index=contact.set_index('Student_ID')
joined_df=student_index.join(contact_index)
print(joined_df.tail(10))

```

# **pivot table in csv:**

```python
#A Pivot Table is a tool used to summarize, organize, and analyze large amounts of data easily.
import pandas as pd
df=pd.read_csv('people-100.csv')
print(df)
print(df.columns)
#Create age column
df['Date of birth']=pd.to_datetime(df['Date of birth']) #here we use to_datetime and Timestamp.today() this functions are used to mention time and date in code
today=pd.Timestamp.today()
df['Age']=((today-df['Date of birth']).dt.days//365)
print(df[['First Name','Sex','Age']].head())
#Basic pivot table
# 1.Average age by gender
pivot1=pd.pivot_table(
    df,
    values='Age',
    index='Sex'
)
print(pivot1)
# 2.pivot table with aggfunc='mean'
pivot2=pd.pivot_table(
    df,
    values='Age',
    index='Sex',
    aggfunc='min'
)
print(pivot2)
# 3.using multiple aggfunc
pivot3=pd.pivot_table(
    df,
    values='Age',
    index='Sex',
    aggfunc=['mean','sum','max']
)
print(pivot3)
# 4.adds overall average
pivot4=pd.pivot_table(
    df,
    values='Age',
    index='Sex',
    aggfunc='mean',
    margins =True
)
print(pivot4)
# 5.groupby jobtitle and gender
#average age by jobtitle and gender
pivot5=pd.pivot_table(
    df,
    values='Age',
    index='Job Title',
    columns='Sex',
    aggfunc='mean',
    margins =True
)
print(pivot5)
# 6.count people by job and gender
pivot6=pd.pivot_table(
    df,
    values='User Id',
    index='Job Title',
    columns='Sex',
    aggfunc='count',
    fill_value=0
)
print(pivot6)
```

# **DATA CLEANING IN EXCEL:**

```python
import pandas as pd
df=pd.read_excel('Data_Cleaning.xlsx')
print(df)
#renaming column
#check correct column name
print(df.columns)
#rename column
df.rename(columns={
    'Student ID':'Student_ID',
    'Student Name ':'Student_Name'
},inplace=True)
print(df.columns)
#Changing Data types
#check data type
print(df.dtypes) 
#Concert Student_ID to String
df['Student_ID'] =  df['Student_ID'].astype(str)
#Convert agr to integer
#First handle missing values
df['Age'] = df['Age'].fillna(0)
df['Age'] = df['Age'].astype(int)
#Check
print(df.dtypes)
     
# 4. Detecting duplicate records
# Find duplicates
duplicates = df.duplicated()
print(duplicates)
#count duplicate
print(df.duplicated().sum())

# 5.dropping duplicate records
#before
print(df.shape)
#remove duplicate
df.drop_duplicates(inplace=True)
#after
print(df.shape)

# 6.String operation
# string operation only works on text column
#Remove extra spaces 
df['Student_Nmae']=df['Student_Name'].str.strip()
#Convert to uppercase
df['Student_Nmae']=df['Student_Name'].str.upper()
print(df['Student_Name'].head())
#covert to title case
df['Student_Nmae']=df['Student_Name'].str.title()
print(df['Student_Name'].head())

# 7.Replace Text
#Replace department name
df['Department']=df[' Department'].replace({
    'CSE':'Computer Science',
    'ECE':'Electronics and communi eng',
    'EEE':'Electrical and electronics',
    'MECH':'Mechanical',
    'IT':'Information Technology'
})
print(df['Department'].head())

# 8.String length
df['Name_length']=df['Student_Name'].str.len()
print(df[['Student_Name','Name_length']].head())

# 9.First letter
df['First_letter']=df['Student_Name'].str.len()
print(df[['Student_Name','First_letter']].head())
```
