# indexing and slicing

```python
import numpy as np
a=np.array([10,20,30,40,50])
print(a[2])
b=np.array([[1,2],
           [4,5]])
#indexing
print(b[1,1])
#1d slicing
print(a[2:5])
print(a[:4])
print(a[:1])
print(a[0:2:1])
print(a[:-2])
print(a[::1])
#2d slicing
print(b[1])
print(b[:,1])
print(a[a>40])
print(a[a<40])
print(a[[0,3]])
print(a[a%2==0])
a[4]=1000
```

## math operation

```python
import numpy as np
a=np.array([[1,2],
           [3,4]])
b=np.array([[3,6],
           [4,5]])
print(a*b)
print(np.dot(a,b))
print(a@b)
print(np.matmul(a,b))
print(a.T)#transpose the array
```

## statistical function

```python
import numpy as np
a=np.array([10,30,40,50,45,60])
print("mean:",np.mean(a))
print("median:",np.median(a))
print("standard division:",np.std(a))
print("max:",np.max(a))
print("min:",np.min(a))
print("percentile:",np.percentile(a,25))
print("var:",(np.var(a)))
b=np.array([1,2,3,4,5])
c=np.array([2,4,6,8,10])
print("coorelation:",np.corrcoef(b,c))
```

## Create 2D Numpy Array:

```python
import numpy as np
a=np.array([[20,40],
           [30,40]])
print("Class Average:",np.mean(a))
high=np.sum(a,axis=1)
print("Score:",high)
print("Max:",max(high))
print("Min:",min(high))
norm=(a-np.min(a))/(np.max(a)-np.min(a))
print("Normalized:",norm)
```

```python
import pandas as pd
#Method1-from dictionary
data={
    'name':['varshini','dharsh','priya'],
    'rollno':[101,102,103]}
df=pd.DataFrame(data)
print(df)
c=pd.Series([1,2,3,4])
d=pd.Series([2.5,3.8,9.0,10.0])
e=pd.Series([True,False,False,True])
df=pd.DataFrame({
    'sno':c,
    'cgpa':d,
    'attend':e})
print(df)
#list of list
data=(['varsh',19,90],
    ['prabha',20,90],
    ['vaish',19,80])
df=pd.DataFrame(data,columns=['name','age','mark'])
print(df)
#list of dictionary
data=(
    {'name':'varsh','course':'aids','grade':'A'},
    {'name':'varshini','course':'aiml','grade':'A++'})
df=pd.DataFrame(data,index=['std1','std2'])
print(df)
#adding extra column 
df['marks']=[100,98]
print(df)
#adding extra row
std3=pd.DataFrame({'name':'dharsh','course':'aids','grade':'A'},
                 index=['std3'])
df=pd.concat([df,std3])
print(df)
#create multiple rows
students=pd.DataFrame([{'name':'nila','course':'aids','grade':'A','marks':80},
                      {'name':'nivan','course':'aids','grade':'A++','marks':90}],index=['std4','std5'])
df=pd.concat([df,students])
print(df)
#column access by braket notation and dot notation
print(df.course)
print(df[['grade','marks']])
# shape attribute
print(df.shape)
#find no of rows
print(df.shape[0])
#find no of column
print(df.shape[1])
#dtype attribute
#used to check data type of all column
print(df.dtypes)
#info
print(df.info())
print(df.describe())
#select specific row
print(df.loc[['std1']])
print(df.iloc[2])
#update value
df.loc['std3','grade']='B++'
print(df)
#delete column
df.drop('grade',axis=1,inplace=True)
print(df)
#delete row
df=df.drop('std5',axis=0)
print(df)
#sorting
df.sort_values('marks',inplace=True)
print(df)
df.sort_values('marks',ascending=False,inplace=True)
print(df)
#stastical function for particular column
print(df['marks'].max())
```
